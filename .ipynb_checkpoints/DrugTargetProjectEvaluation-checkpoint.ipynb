compare simple weighted profile and hubnesaware one on each data sets

```python
# rewrite the evaluation to show curve
%matplotlib inline
def Evaluation(Interactions,NewInteractions):

    NumOfDrugs = Interactions.shape[0];

    NumOfTargets = Interactions.shape[1];

    TruelLabels = []

    Scores = []

    for i in range(0,NumOfDrugs):
        for j in range(0,NumOfTargets):
            
            TruelLabels.append(Interactions.iloc[i,j]);

            Score = NewInteractions.iloc[i,j];
            
            if Score!=Score:
                Scores.append(0)
            else:
                Scores.append(Score)
        
        prec, rec, thr = precision_recall_curve(TruelLabels, Scores)
        
        aupr_val = auc(rec, prec)

        fpr, tpr, thr = roc_curve(TruelLabels, Scores)

        auc_val = auc(fpr, tpr)
        
        plt.plot(fpr, tpr)
        plt.plot(prec, rec)
        plt.show()
        
        return auc_val,aupr_val
```

```python
#Kinase Data Sets
SimpleResultKinase = SimpleWeightedProfile.WeightedProfile(DDSimilarityKinase,TTSimilarityKinase,InteractionsKinase,2)
HubnessAwareResultKinas = HubnessAware.WeightedProfile(DDSimilarityKinase,TTSimilarityKinase,InteractionsKinase,2)
```

```python
AUC_SimpleKinase , AUPR_SimpleKinase = Evaluation(InteractionsKinase,SimpleResultKinase)
AUC_HubnessKinase , AUPR_HubnessKinase = Evaluation(InteractionsKinase,HubnessAwareResultKinas)
print(AUC_SimpleKinase)
print(AUC_HubnessKinase)
print(AUPR_SimpleKinase)
print(AUPR_HubnessKinase)
```

```python
#GPCR Data Sets
SimpleResultGPCR = SimpleWeightedProfile.WeightedProfile(DDSimilarityGPCR,TTSimilarityGPCR,InteractionsGPCR,2)
HubnessAwareResultGPCR = HubnessAware.WeightedProfile(DDSimilarityGPCR,TTSimilarityGPCR,InteractionsGPCR,2)
```

```python
AUC_SimpleGPCR , AUPR_SimpleGPCR = Evaluation(InteractionsNR,SimpleResultGPCR)
AUC_HubnessGPCR , AUPR_HubnessGPCR = Evaluation(InteractionsNR,HubnessAwareResultGPCR)
print(AUC_SimpleGPCR)
print(AUC_HubnessGPCR)
print(AUPR_SimpleGPCR)
print(AUPR_HubnessGPCR)
```

hubnessaware weighted profile NumOfNeighbours choose

```python
n_range = range (1,5)
n_auc = []
n_aupr = []
for n in n_range:
    HubnessAwareResultKinas_n = HubnessAware.WeightedProfile(DDSimilarityKinase,TTSimilarityKinase,InteractionsKinase,n)
    auc_n, aupr_n = Evaluation(InteractionsKinase,HubnessAwareResultKinas_n)
    n_auc.append(auc_n)
    n_aupr.append(aupr_n)
```

```python
plt.plot(n_range,n_auc)
plt.plot(n_range,n_aupr)
plt.show()
```
